```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, when, trim, lower, count, sum, avg, min, max
)
from pyspark.sql.types import (
    StructType, StructField, IntegerType,
    StringType, DoubleType
)
spark = SparkSession.builder \
    .appName("SparkDataFrameDataProcessing") \
    .master("local[*]") \
    .getOrCreate()

df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("C:\\Users\\VIVOBOOK PRO\\Desktop\\superstore1.csv")



print("Original Dataset")
df.show(5)

print("Schema:")
df.printSchema()

```

```python
print("Dataset Schema:")
df.printSchema()

print("Number of Rows:", df.count())
print("Number of Columns:", len(df.columns))
```

```python
df_cleaned = df.dropDuplicates(
    ["Order ID", "Customer Name", "Category"]
)
print("Data After filteration")
df_cleaned.show()
```

```python
for column in df_cleaned.columns:
    print(
        column,
        df_cleaned.filter(col(column).isNull()).count()
    )
```

```python
df_deduplicated = df.drop_duplicates(["Order ID", "Customer Name", "Category"])
print("Data after first filteration")
print("data after removing the duplicates record from the original data")
print(df_deduplicated.show())
```

```python
from pyspark.sql.functions import col, when

df = df.withColumn(
    "Sales",
    when(
        col("Sales").isin("null", "NULL", ""),
        None
    ).otherwise(
        col("Sales").cast("double")
    )
)
```

```python
from pyspark.sql.functions import avg

west_sales = df \
    .filter(col("Region") == "West") \
    .groupBy("Category") \
    .agg(
        avg("Sales").alias("average_sales")
    )

west_sales.show()
```

```python
df_without_nulls = df.na.drop()
df_without_nulls.show()
```

```python
df_filled = df.na.fill({
    "Customer Name": "Unknown"
})

df_filled.show()
```

```python
city_counts = df \
    .groupBy("City") \
    .agg(
        count("*").alias("total_count")
    ) \
    .filter(
        col("total_count") > 100
    )

city_counts.show()
```

```python
df_without_segment = df.drop("Segment")
print("Data after removing the column")
df_without_segment.show()
```

```python
df_renamed = df.withColumnRenamed(
    "Customer Name",
    "Customer_Name"
)
print(df_renamed.show())
```

```python
filtered_data = df.filter(
    (col("Region") == "West") &
    (col("Sales") > 100)
)
print("Data after filteration")
filtered_data.show()
```

```python
df_cleaned = df.na.fill({
    "Sales": 0
})
```

```python
df_cleaned.agg(
    avg("Sales").alias("average_sales")
).show()
```

```python
from pyspark.sql.functions import lit, to_timestamp, col

df_timestamp = df.withColumn(
    "raw_timestamp",
    to_timestamp(
        lit("2025-01-01"),
        "yyyy-MM-dd"
    )
)

df_timestamp = df_timestamp.withColumnRenamed(
    "raw_timestamp",
    "event_time"
)

df_timestamp.select(
    "Order ID",
    "event_time"
).show()
```

```python
category_sales = df \
    .groupBy("Category") \
    .agg(
        sum("Sales").alias("total_sales")
    )
category_sales.show()
```

```python
cleaned_df = df.filter(
    col("Customer Name").isNotNull() &
    (trim(col("Customer Name")) != "")
)

cleaned_df.show()
```

```python
sales_statistics = df.agg(
    min("Sales").alias("minimum_sales"),
    max("Sales").alias("maximum_sales"),
    avg("Sales").alias("average_sales")
)

sales_statistics.show()
```

```python
final_result = df \
    .dropDuplicates(
        ["Order ID", "Customer Name", "Category"]
    ) \
    .na.fill({
        "Sales": 0
    }) \
    .groupBy("Region") \
    .agg(
        sum("Sales").alias("total_revenue")
    )

final_result.show()
```
