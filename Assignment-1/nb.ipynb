```python
import pandas as pd

# Load the dataset
df = pd.read_csv("backpack1.csv")

print("=" * 50)
print("FIRST 5 ROWS")
print(df.head())

print("\n" + "=" * 50)
print("LAST 5 ROWS")
print(df.tail())

print("\n" + "=" * 50)
print("DATASET SHAPE")
print(df.shape)

print("\n" + "=" * 50)
print("COLUMN NAMES")
print(df.columns.tolist())

print("\n" + "=" * 50)
print("DATA TYPES")
print(df.dtypes)

print("\n" + "=" * 50)
print("MISSING VALUES")
print(df.isnull().sum())

# Handle Missing Values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].mean())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Remove Duplicates
duplicates = df.duplicated().sum()
print(f"\nDuplicate Rows Found: {duplicates}")

df = df.drop_duplicates()
print(f"Rows after removing duplicates: {len(df)}")

# Select Columns
print("\nSelected Columns:")
print(df[["price", "quantity"]].head())

# Filter Rows
filtered_df = df[df["price"] > 100]

print("\nFiltered Rows (price > 100):")
print(filtered_df.head())

# Create Derived Column
df["total_amount"] = df["price"] * df["quantity"]

print("\nDataset with Total Amount:")
print(df.head())

# Save Cleaned Dataset
df.to_csv("cleaned_sales_data.csv", index=False)

print("\nCleaned dataset saved as 'cleaned_sales_data.csv'")

# Summary
print("\n" + "=" * 50)
print("SUMMARY")
print(f"Original Columns: {len(df.columns)-1}")
print(f"Final Columns: {len(df.columns)}")
print(f"Total Records: {len(df)}")
print("Added Column: total_amount")
print("Duplicates Removed Successfully")
print("Missing Values Handled Successfully")
```
